```python
import pandas as pd
import numpy as np

np.random.seed(42)
jumlah_data = 1000

nama_atlet = ['Atlet_' + str(i) for i in range(1, jumlah_data + 1)]

data = pd.DataFrame({
    'Nama_Atlet': nama_atlet,
    'Usia': np.random.randint(15, 35, size=jumlah_data),
    'Jenis_Kelamin': np.random.choice(['Laki-laki', 'Perempuan'], size=jumlah_data),
    'Cabang_Atletik': np.random.choice(['Lari 100m', 'Lari 400m', 'Lompat Jauh', 'Lempar Lembing', 'Lari Marathon'], size=jumlah_data),
    'Waktu_Latihan_per_Hari': np.round(np.random.uniform(1, 4, size=jumlah_data), 1),
    'Frekuensi_Latihan_Mingguan': np.random.randint(3, 7, size=jumlah_data),
    'Catatan_Waktu_Terbaik': np.round(np.random.uniform(10.0, 300.0, size=jumlah_data), 2),  # detik
    'Tinggi_Badan_cm': np.random.randint(150, 200, size=jumlah_data),
    'Berat_Badan_kg': np.random.randint(45, 90, size=jumlah_data),
    'Lolos_Seleksi': np.random.choice([0, 1], size=jumlah_data, p=[0.5, 0.5])
})

print("Contoh data atlet:")
print(data.head())

print("\nStatistik deskriptif:")
print(data.describe(include='all'))

```

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for col in ['Jenis_Kelamin', 'Cabang_Atletik']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

data.to_csv('data_atlet_encoded.csv', index=False)
```
