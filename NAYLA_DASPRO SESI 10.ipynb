```python
import pandas as pd
import numpy as np

np.random.seed(42)

jumlah_data = 1000

data = pd.DataFrame({
    'Usia': np.random.randint(18, 65, size=jumlah_data),
    'Pendapatan': np.random.normal(5000000, 1500000, size=jumlah_data).astype(int),
    'Jenis_Kelamin': np.random.choice(['Laki-laki', 'Perempuan'], size=jumlah_data),
    'Status_Pernikahan': np.random.choice(['Lajang', 'Menikah', 'Bercerai'], size=jumlah_data),
    'Tingkat_Pendidikan': np.random.choice(['SMA', 'S1', 'S2', 'S3'], size=jumlah_data),
    'Status_Pekerjaan': np.random.choice(['Bekerja', 'Pengangguran', 'Wirausaha', 'Mahasiswa'], size=jumlah_data),
    'Skor_Kredit': np.random.randint(300, 850, size=jumlah_data),
    'Jumlah_Rekening_Bank': np.random.randint(1, 6, size=jumlah_data),
    'Jumlah_Pinjaman': np.random.normal(15000000, 5000000, size=jumlah_data).astype(int),
    'Gagal_Bayar': np.random.choice([0, 1], size=jumlah_data, p=[0.85, 0.15])  # 0 = Tidak, 1 = Ya
})

print("Contoh data:")
print(data.head())

print("\nStatistik deskriptif:")
print(data.describe(include='all'))
```
